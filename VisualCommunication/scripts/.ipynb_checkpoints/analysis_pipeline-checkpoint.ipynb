# Lets plot some data from our example, looking at head position movement
Below we plot the head movement from the example video above. We simply plot the vertical position over time (frames). We also apply some filtering to reduce noise-related jitter. We use a savitsky golai filter for this, with a span of x, and a order of y, such that savgol_filter(timeseries, x, y).

```python
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt

# lets just show the nose data
nosex = MTface[nose[0][0]]
nosey = MTface[nose[0][1]]
nosez = MTface[nose[0][2]]
  # smooth the nose with savitsky golay filter
nosey = nosey*-1 # invert the y axis as down is up in the data
nosey = savgol_filter(nosey, 26, 3)

# now we can plot the nose vertical position
plt.plot(nosey)
plt.title('Vertical position nose')
plt.xlabel('frames')
    #add a custom v line, initialized at 10
#plt.axvline(x=10, color='r', linestyle='--')
plt.show()
```

# Sanity checking
It is really important to check your measures with the video. For example, are your smoothing the data too much. Are you accounting for the right type of movement (e.g., maybe you should center your data around a particular reference point). We will do this sanity checking continuously throughout the course. Here we do it by plotting the vertical position continuously with the video. Does it correspond to the video?

```python
# lets make a video where this plot is embedded on the left upper corner as a small plot with the vertical line moving
# we loop through each frame of the video using opencv, and add the plot to the video
# we also make a function as we are going to add this plot in a video to assess the velocity of the nose next to the video

# first we need to load in the video
import cv2
import numpy as np
from moviepy.editor import VideoFileClip
from IPython.display import Video
import matplotlib.pyplot as plt 


# function to plot the velocity of the nose
def plot_velocity(ts, titlefig, trial_name, threshold=10):
    plt.plot(ts)
    plt.title(titlefig)
    plt.axvline(x=threshold, color='r', linestyle='--')
    plt.savefig('../temp/' + trial_name + '_pos.png')
    plt.close()
    imloc = '../temp/' + trial_name + '_pos.png'
    return imloc
    
# Load in the video
cap = cv2.VideoCapture(masked_video_fol + 'alt' + select_example + '.mp4')
# get the frame width and height and make it bigger
frame_width = int(cap.get(3))*4
frame_height = int(cap.get(4))*4
# define the codec and create a VideoWriter object
out = cv2.VideoWriter('../augmented_videos/'+'output_withdata.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 25, (frame_width,frame_height))

# loop through the video and add the plot to the video on left upper corner in small inset
frame_number = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        frame = cv2.resize(frame, (int(frame_width), int(frame_height)))
        # add the plot to the video
        imloc = plot_velocity(nosey, 'vertical position nose', select_example, frame_number)
        img = cv2.imread(imloc)
        img = cv2.resize(img, (int(frame_width)//4, int(frame_height)//4))
        frame[:img.shape[0], :img.shape[1]] = img
        out.write(frame)
        frame_number += 1
    else:
        break

# Release everything if job is finished
cap.release()
out.release()

# also rerender it with moveiepy
from moviepy.editor import VideoFileClip
clip = VideoFileClip('../augmented_videos/'+'output_withdata.avi')
clip.write_videofile('../augmented_videos/'+'output_withdata.mp4')

# show a video of the exmaple
width = int(frame_width//4)  # Adjust this value as needed
height = int(frame_width//4) # Adjust this value as needed
Video('../augmented_videos/'+'output_withdata.mp4', width=width, height=height)
```

# Compute for the whole data
Now that we have done some sanity checking, and if we trust our operationalization of things, we can start to process all the trials and extract relevant kinematic parameters from the timeseries. We are extracting the following below and are adding this to our flat dataset:

* max amplitude of head nod
* max vertical velocity

This we add into our meta dummy dataset, which will allow you to do comparisons between groups later on.

```python
# now lets loop through the meta data, enrich the columns with kinematic summary data 
# lets loop through the meta
for row in range(len(meta)):
    # get the trial name
    trial_name = meta['trial_name'][row]
    # Load in the facial landmarks data
    MTface = pd.read_csv(processed_data_fol + trial_name + "_face.csv")
    # lets just show the nose data
    nosex = MTface[nose[0][0]]
    nosey = MTface[nose[0][1]]
    nosez = MTface[nose[0][2]]
    # smooth the nose with savitsky golay filter
    nosey = nosey*-1 # invert the y axis as down is up in the data
    nosey = savgol_filter(nosey, 26, 3)
        # calculate the nose velocity
    nose_vel = np.diff(nosey)
        
    # save the max and min of the nose y position
    meta.at[row, 'nosey_max'] = nosey.max()-nosey.min()
    # save the max velocity (in any direction)
    nose_velabs = np.abs(nose_vel)
    meta.at[row, 'nosey_vel_max'] = nose_velabs.max()
 # save the meta as an enriched file
meta.to_csv("../meta/meta_dummy_enriched.csv", index=False)

print('this is what the enriched "flat" dataset now looks like after extracting the kinematics')
meta.head()
```
